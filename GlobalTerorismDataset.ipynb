```python
#Importing Libraries
import numpy as np
import matplotlib.pyplot as plt    
import pandas as pd             
import seaborn as sns       
%matplotlib inline
```

```python
#Reading Data from csv file
terr=pd.read_csv('C:/Users/USER/OneDrive/Desktop/LGM-Task-3/globalterrorismdb_0718dist.csv',low_memory=False)
terr.head()    
```

```python
terr.tail()
```

```python
terr.rename(columns={"iyear":"Year","imonth":"Month","iday":"Day","extended":"Extended","provstate":"State","country_txt":"Country",
    "region_txt":"Region","attacktype1_txt":"Attack Type","target1":"Target","nkill":"Killed","nwound":"Wounded","summary":"Summary","gname":"Group",
    "weaptype1_txt":"Weapon Type","motive":"Motive","latitude":"Latitude","longitude":"Longitude","city":"City",
    "eventid":"Event","resolution":"Resolution"},inplace=True)
```

```python
terr=terr[["Year","Month","Day","Extended","State","Country","Region","City","Latitude","Longitude","Attack Type",
        "Target","Wounded","Killed","Summary","Group","Motive","Weapon Type"]]
```

```python
terr.head(20)
```

```python
#Adding New Column named Casaulty
terr["Killed"]=terr["Killed"].fillna(0)
terr["Wounded"]=terr["Wounded"].fillna(0)
terr["Casualty"]=terr["Killed"]+terr["Wounded"]
```

```python
terr.info()
```

```python
terr.isnull().sum()
```

```python
terr.describe()
```

#### __Countries with Most Number of Terrorist Attacks__

```python
terrco=terr["Country"].value_counts().head(20)
print("Top 20 Countries with most number of Terrorist Attacks:- \n",terrco)
```

```python
plt.figure(figsize=(18,9))
terrco.plot(kind="bar",color="Blue",fontsize=12,label="Attacks")
plt.xlabel("Country",fontsize=16)
plt.ylabel("Number of Attacks ->",fontsize=16)
plt.title("Top 20 Countries with most number of Terrorist Attacks",fontsize=15)
plt.legend(loc="upper right")
plt.show()
```

#### __States/Provinces with Most Number of Terrorist Attacks__

```python
terrst=terr["State"].value_counts().head(20)
print("Top 20 States/Provinces with most number of Terrorist Attacks:- \n",terrst)
```

```python
plt.figure(figsize=(18,9))
terrst.plot(kind="bar",color="DarkGreen",fontsize=12,label="Attacks")
plt.xlabel("State",fontsize=16)
plt.ylabel("Number of Attacks ->",fontsize=16)
plt.title("Top 20 States/Provinces with most number of Terrorist Attacks",fontsize=15)
plt.legend(loc="upper right")
plt.show()
```

#### __Regions with Most Number of Terrorist Attacks__

```python
terrre=terr["Region"].value_counts().head(20)
print("Top 20 Regions with most number of Terrorist Attacks:- \n",terrre)
```

```python
plt.figure(figsize=(20,10))
terrre.plot(kind="bar",color='red',fontsize=12,label="Attacks")
plt.xlabel("Region",fontsize=16)
plt.ylabel("Number of Attacks ->",fontsize=16)
plt.title("Top 20 Regions with most number of Terrorist Attacks",fontsize=15)
plt.legend(loc="upper right")
plt.show()
```

```python
terrreg=pd.crosstab(terr.Year,terr.Region)
terrreg.plot(kind="area",figsize=(20,10),stacked="False",alpha=0.5)
plt.xlabel("Year",fontsize=16)
plt.ylabel("Number of Attacks->",fontsize=16)
plt.title("Attacks on Regions",fontsize=15)
plt.show()
```
